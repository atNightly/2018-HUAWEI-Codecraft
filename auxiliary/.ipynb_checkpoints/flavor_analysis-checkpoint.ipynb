```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date
%matplotlib inline
```

```python
data = '../data/用例示例/TrainData_2015.1.1_2015.2.19.txt'
df = pd.read_csv(data, delimiter='\t', header=None, 
                 names=['vm-id', 'vm-name', 'create_at'],
                parse_dates=['create_at'])
df['create_date'] = df['create_at'].apply(lambda x: x.date())
```

```python
interval = df.create_date.unique()
```

```python
flavors_complete = []
for flavor in df['vm-name'].unique():
    f = df[df['vm-name'] == flavor].groupby('create_date')['vm-name'].count()
    for intval in interval:
        if intval not in f:
            f = f.append(pd.Series({intval: 0})).sort_index()
            frame = f.to_frame().rename(columns={0: 'count'})
            frame['cumsum'] = frame['count'].cumsum()

    flavors_complete.append((flavor, frame))
```

```python
plt.figure(figsize=(16, 12))
for f in flavors_complete:  
    plt.plot(f[1].index, f[1]['count'], 'o--', label=f[0])
plt.legend()
```

```python
plt.figure(figsize=(16, 12))
for f in flavors_complete:  
    plt.plot(f[1].index, f[1]['cumsum'], 'o-', label=f[0])
plt.legend()
```
